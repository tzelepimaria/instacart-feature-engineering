```python
#10% of dataset
orders = orders.loc[orders.user_id.isin(orders.user_id.drop_duplicates().sample(frac=0.1, random_state=25))]

#merge to create op
op = orders.merge(order_products_prior, on='order_id', how='inner')
op.head(10)
```

```python
#mean cart position per product
product_cart_position = op.groupby('product_id')['add_to_cart_order']\
.mean().to_frame('p_mean_cart_position') \
.reset_index()

#add name of products
product_cart_position = product_cart_position.merge(products[['product_id', 'product_name']],on='product_id',how='left')

product_cart_position = product_cart_position.sort_values(by='p_mean_cart_position',ascending=True)

product_cart_position.head(20)


```
